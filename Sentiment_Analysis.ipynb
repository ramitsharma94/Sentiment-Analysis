```python
### Importing the libraries 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import string 

import spacy
import nltk
from nltk.stem import WordNetLemmatizer
import en_core_web_sm
from spacy.lang.en.stop_words import STOP_WORDS

nlp = en_core_web_sm.load()
stopwords = list(STOP_WORDS)

punctuations = string.punctuation
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
### Code to scrap the headlines and summary from The Print website

import requests 
from bs4 import BeautifulSoup
import json
import datetime

headlines = []
author_and_dates = []
summaries = []

for i in range(1,1391):    
    url = f"https://theprint.in/category/india/page/{i}/"
    print(url)
    r = requests.get(url)
    r_html = r.text
    soup = BeautifulSoup(r_html, "html.parser")
    scripts = soup.find_all("div", class_="item-details")
    for script in scripts:
        if((script.find("div", class_="td-excerpt")) and script.find("div", class_ = "td-module-meta-info")):
#             print("New\n", script)
            headline = script.find("h3").text
#             print(headline)
            author = script.find("span").text
#             print(author)
            date = script.find("time").text
#             print(date)
            author_date = author + date
            summary = script.find("div", class_ = "td-excerpt").text
            headlines.append(headline)
            author_and_dates.append(author_date)
            summaries.append(summary)
        continue
        
        
df = pd.DataFrame({"Headlines": headlines, 
                  "Author_Dates": author_and_dates,
                  "Summaries": summaries} )
df.head()
```

```python
### Function to clean the text which we will do later in this notebook 

def clean_text(sent):
    sentence = sent.strip()
    sentence = sentence.lower()
    tokens = nltk.word_tokenize(sentence)
    tokens = [token for token in tokens if not (token in punctuations or token in stopwords)]
    
    lemmatized_token = [wordnet_lemmatizer.lemmatize(token) for token in tokens]
    new_sentence = " ".join(lemmatized_token)
    return new_sentence
```

```python
# get topics with their terms and weights 

def get_topics_term_weights(weights, feature_names):
    feature_names = np.array(feature_names)
    sorted_indices = np.array([list(row[::-1]) for row in np.argsort(np.abs(weights))])
    sorted_weights = np.array([list(wt[index]) for wt, index in zip(weights, sorted_indices)])
    sorted_terms = np.array([list(feature_names[row]) for row in sorted_indices])
    
    topics = [np.vstack((terms.T, term_weights.T)).T for terms, term_weights in zip(sorted_terms, sorted_weights)]
    return topics


topics = get_topics_term_weights(nmf_h.components_, tfidfvect_h.get_feature_names())
topics
```

```python
def print_topics_udf(topics, total_topics = 1, 
                    weight_threshold = 0.0001,
                    display_weights = False,
                    num_terms = None):
    
    for index in range(total_topics):
        topic = topics[index]
        topic = [(term, float(wt)) for term, wt in topic]
        
        topic = [(word, round(wt,2)) for word, wt in topic if abs(wt)>= weight_threshold]
        
        if(display_weights):
            print("Topic #: {} with weights".format(index+1))
            print(topic[:num_terms]) if num_terms else topic
        else:
            print("Topic # {} without weights".format(index+1))
            tw = [term for term, wt in topic]
            print(tw[:num_terms]) if num_terms else tw
            
print_topics_udf(topics, total_topics=2, num_terms=30, display_weights= True)
```

```python
def get_topics_udf(topics, total_topics = 2, 
                  weight_threshold = 0.0001,
                  num_terms = None):
    
    topic_terms = []

    for index in range(total_topics):
        print("topics[{}]".format(index))
        topic = topics[index]
        print(topic)
        topic = [(term, float(wt)) for term, wt in topic]
        print(len(topic))
        topic = [(word, round(wt, 2)) for word, wt in topic if abs(wt)>= weight_threshold]
        print(len(topic))
        topic_terms.append(topic[:num_terms] if num_terms else topic)
    return topic_terms
                
        
def get_terms_size(topic_display_list_item):
    
    terms = []
    sizes = []
    
    for term,size in topic_display_list_item:
        terms.append(term)
        sizes.append(size)
    return terms, sizes

topics_display_list = get_topics_udf(topics, total_topics = 2, 
                  num_terms = 30)
```
